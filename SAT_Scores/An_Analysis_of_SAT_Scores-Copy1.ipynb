## Visualizing the Data

#### Distribution Plots of Verbal and Math Scores and Participation Rate

```python
fig = plt.figure()
sns.distplot(satdf.Rate, bins=15, kde=True, color='Green')
plt.ylabel('Relative Frequency')
plt.title('Distribution of Participation Rate of State');
```

```python
fig, (ax1, ax2) = plt.subplots(2, sharex=True, figsize=(6,6))
sns.distplot(satdf.Math, bins=15, kde=True, color='Red', ax=ax1)
sns.distplot(satdf.Verbal, bins=15, kde=True, ax=ax2)
ax2.set_ylabel('Relative Frequency')
ax1.set_ylabel('Relative Frequency')
plt.title("Distribution of Math and Verbal Scores");
```

The typical assumption for a data distribution is that it will be normally distributed. A visual check already shows that the distribution for the Participation Rate is not a normal distribution and the Math and Verbal are closer to normal distributions, so a statistical normality test would be useful to double check their normality.

#### Testing the normality of distributions for each variable - Participation Rate, Verbal Score, Math Score

```python
stats.normaltest(satdf['Rate'])
```

```python
stats.shapiro(satdf[['Rate']])
```

The P-values for the Chi-Squared Goodness of fit test and the Shapiro Wilks test show that the Participation Rates are not normally distributed and we can reject the null hypothesis.

```python
stats.normaltest(satdf['Math'])
```

```python
stats.shapiro(satdf['Math'])
```

The P-values for the Chi-Squared Goodness of fit test and the Shapiro Wilks test show that there is not enough evidence to reject the null hypothesis that Math scores are normally distributed. A brief look at the distribution histogram and the boxplot gives one the idea that Math scores are the most normally distributed.

```python
stats.normaltest(satdf['Verbal'])
```

```python
stats.shapiro(satdf['Verbal'])
```

The P-values for the Chi-Squared Goodness of fit test and the Shapiro Wilks test show that Verbal scores are not normally distributed and we can reject the null hypothesis.

#### Boxplots

```python
fig = plt.figure(figsize=(6,6))
sns.boxplot(satdf[['Math','Verbal']], palette="husl")
plt.ylabel('Score')
plt.title('Box Plots of Verbal and Math Scores');
```

From our previous look at the min and maxes, we could assume the verbal scores are more tightly packed and the whiskers on the math scores are longer than on the verbal scores, even though the interquartile range for the verbal scores is larger than in math scores.

```python
fig = plt.figure(figsize=(6,3))
sns.boxplot(satdf['Rate'], palette='husl')
plt.title('Box Plot of Partipication Rate of State');
```

The interquartile range is quite large on the participation rate, from under 10% to a bit under 65% participation rate, which indicates a fairly disperse distribution of participation rates by state, typically lower than 50% on average.

```python
fig = plt.figure(figsize=(6,6))
sns.regplot(satdf['Rate'],satdf['Math'], color='red')
sns.regplot(satdf['Rate'],satdf['Verbal'], color='green')
plt.ylabel('Score')
plt.legend(['Math','Verbal'])
plt.title('Scores vs. Participation Rate');
```

```python
fig = plt.figure(figsize=(6,6))
sns.regplot(satdf['Math'],satdf['Verbal'])
plt.title('Math Score vs. Verbal Score');
```

The plots seem to suggest that verbal and math scores seem to be positively correlated whilst both Math and Verbal scores seem to be negatively correlated to participation rate. 

This would indicate that the higher the participation rate of students in a particular state the lower the scores for Verbal and Math scores would be. This suggests that in States with very low participation rates, the students who decide to sit for the SAT tend to have studied harder and perform better than the average student in higher participation rate states.

Let's check the correlation:

```python
satdf.corr(method='pearson')
```

The correlation matrix seems to prove what one expects, Verbal and Math scores are correlated strongly positively and both are very strongly negatively correlated to Partipication Rate.

### Conclusions:

As one would expect Math and Verbal scores are highly correlated, typically performing poorly in one would correspond to a poor performance in the other and this seems to be proven by the strong correlation between the two.

A little more surprising is that participation rate of the State has such a strong negative correlation with scores in both Verbal and Math, which indicates that low participation states are biased with only the strongest performing participating in the SAT. 

It would be useful to have other features in examining what affects scores and participation in the examination itself, for example median household income would be a useful metric or the number of colleges per capita would also be interesting to explore how it affects both scores and participation rates. If limited to the dataset we are working with, you could engineer a feature like weighted scores based on participation rate for the state and of course it would be simple to create a feature with the combined Verbal and Math scores as well or the difference between the two scores.
